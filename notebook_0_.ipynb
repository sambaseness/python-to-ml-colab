# Jour 0 : Découverte de Google Colab et des Notebooks Python

Bienvenue dans cet environnement de travail interactif ! Avant de commencer à coder, il est essentiel de comprendre l'outil que nous allons utiliser.

## 1. Qu'est-ce qu'un Notebook Python ?

Un notebook (carnet en français) est un document interactif qui contient deux types de cellules principales :

1. **Les cellules de texte (Markdown)** : Comme celle que vous lisez actuellement. Elles servent à expliquer la théorie, mettre des images ou des consignes.
2. **Les cellules de code (Python)** : Elles contiennent du code exécutable. Lorsque vous les exécutez, le résultat s'affiche juste en dessous.

C'est l'outil parfait pour apprendre pas à pas, car on peut exécuter le code petit à petit et voir le résultat immédiatement.

## 2. Qu'est-ce que Google Colab ?

Google Colaboratory (ou Colab) est un service gratuit proposé par Google. C'est un environnement de notebooks hébergé dans le cloud.

**Ses avantages pour les débutants :**
- **Zéro installation** : Pas besoin d'installer Python ou Jupyter sur votre ordinateur. Tout se passe dans votre navigateur web.
- **Gratuit** : Vous avez accès à des ressources de calcul gratuitement.
- **Sauvegarde automatique** : Vos fichiers sont sauvegardés sur votre Google Drive.
- **Partage facile** : Comme un document Google Docs, vous pouvez partager vos notebooks.

## 3. Comment accéder à Google Colab ?

1. Ouvrez votre navigateur web (Chrome, Firefox, etc.).
2. Allez à l'adresse : https://colab.research.google.com
3. Connectez-vous avec votre compte Google (Gmail).
4. Une fenêtre pop-up s'ouvre. Cliquez sur **Nouveau notebook** (ou allez dans Fichier > Nouveau notebook).

Et voilà ! Vous êtes prêt à coder.
